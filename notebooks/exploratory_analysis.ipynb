```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed dataset
df = pd.read_csv("../data/processed_orders.csv", parse_dates=["OrderDate","ShipDate","DeliveryDate"])

# Preview
print(df.head())

# Delay distribution
plt.hist(df["DeliveryDelay_Days"], bins=30, color="skyblue")
plt.title("Delivery Delay Distribution")
plt.xlabel("Days")
plt.ylabel("Frequency")
plt.show()

# Group by courier
print(df.groupby("Courier")["IsDelayed"].mean())
```
